```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

```python
x = ['Mon', 'Tues', 'Wed', 'Thus', 'Fri']
y = [10,15,7,20,12]

plt.plot(x, y) # plot the data
plt.title('Sales Over A Week') # Adding titile
plt.xlabel('Days') # Label for the x-axis
plt.ylabel('Sales') # Label for the y-axis
plt.show() # Display the plot
```

```python
plt.plot(x, y, color="red", linestyle="dotted", linewidth=2, marker=">", label="label")
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales Per Week')
plt.legend()
plt.grid()
plt.xlim(1,5)
plt.ylim(1,25)
plt.yticks([7, 20], ['minimum', 'maximum'])
plt.show()
```

```python
months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label='2026 Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales Per Month')
plt.title('Monthly Sales Data Report')
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(1,4)
plt.xticks([1, 2, 3, 4], ['M1', 'M2', 'M3', 'M4'])
plt.ylim(800, 2000)
plt.show()
```

#### Bar Chart

```python
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.bar(product, sales, color='orange', label='Sales 2026')
#plt.barh(product, sales, color='orange', label='Sales 2026')   for horizontal

plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

#### Pie Chart

```python
regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution by Region')
plt.show()
```

#### Histogram

```python
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color='blue', edgecolor='white')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')
plt.show()
```

#### Scatter Plot

```python
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies, exam_scores, color='green', marker='^', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship b/w Study and Exam Score')
plt.legend()
plt.grid(color='gray', )
plt.show()
```

#### Compare Two Classes

```python
plt.scatter([1,2,3],[50,55,60],color='blue',label='Class A') #group 1
plt.scatter([1,2,3],[45,50,62],color='red',label='Class B') #group 2

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparision of Two Classes')
plt.legend()
plt.grid(True)
plt.show()
```

## Sub Plot

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.subplot(1, 2, 1) # 1 row, 2 column, 1st subplot
plt.plot(x, y)
plt.title('Line Chart')

plt.subplot(1, 2, 2) # 1 row, 2 col, 2nd subplot
plt.bar(x, y)
plt.title('Bar Chart')
```

##### Advance way of sub ploting

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax[0].plot(x, y, color='blue')
ax[0].set_title('Line Plot')

ax[1].bar(x, y, color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line and Bar Charts')

plt.tight_layout() # auto adjusment
plt.show()
```
