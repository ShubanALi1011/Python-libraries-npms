```python
import numpy as np
import pandas as pd;
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data set
tips = sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
sns.set_theme('notebook')
sns.set_style("whitegrid")
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips
)
plt.show()
```

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    data=tips
)
plt.show()
```

```python
# histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g", hue="sex", multiple="stack" )
```

```python
# regression plt
sns.regplot(data=penguins, 
            x="body_mass_g", 
            y="flipper_length_mm",
            color="green",)
plt.show()
```

```python
# line plot
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="sex", style="sex")
```

```python
# joint plot
sns.jointplot(data=penguins, 
              x="body_mass_g", 
              y="flipper_length_mm",
              hue="sex")
```

```python
# joint plot
sns.jointplot(data=penguins, 
              x="body_mass_g", 
              y="flipper_length_mm",
              kind="reg")
```

```python
# bar plot
sns.barplot(data=penguins, x="species", y="body_mass_g", hue ="sex", palette=['blue', 'orange'], estimator=np.mean) 
```

```python
# count plot
sns.countplot(data=penguins, x="species", hue="island", color="pink")
```

```python
# box plot
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue ="sex")
```

```python
# violin plot
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue ="sex", palette="Set2", split=True)
```

```python
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue ="sex", palette="Set2", split=True)
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue  ="sex", dodge=True, color="k", alpha=0.6)
```

```python
# kde plot
sns.kdeplot(data=penguins, x="body_mass_g", hue="sex")
```

```python
# Heatmap
penguins.head()
```

```python
columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
sns.heatmap(penguins[columns].corr(), annot=True, cmap="YlGnBu")
```

```python
# Rug plot
sns.rugplot(data=penguins, x="body_mass_g", hue="sex")
```

```python
# pair plot
sns.pairplot(penguins, hue="sex", diag_kind="hist") 
```

```python
# pair grid
g = sns.PairGrid(penguins, hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
